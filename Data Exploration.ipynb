```python
# Performing two-way ANOVA
model = ols('HeartDisease ~ C(Smoking) + C(AlcoholDrinking) + C(HeartDisease):C(Smoking)', data=data).fit()
result = sm.stats.anova_lm(model, type=2)
print(result)
```

```python
# conducting the Shapiro-Wilk Test
print(shapiro(Smoking))
print(shapiro(AlcoholDrinking))
print(shapiro(BMI))
print(shapiro(Stroke))
print(shapiro(PhysicalHealth))
print(shapiro(MentalHealth))
print(shapiro(DiffWalking))
print(shapiro(Sex))
print(shapiro(Diabetic))
print(shapiro(PhysicalActivity))
print(shapiro(SleepTime))
print(shapiro(Asthma))
print(shapiro(KidneyDisease))
print(shapiro(SkinCancer))
```
